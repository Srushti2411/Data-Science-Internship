```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
x = np.array([12,23,34,45,49,56,61,68,71,78])
y = np.array([23,31,38,43,49,56,65,71,79,85])
z = np.array([98,89,86,76,71,68,61,54,49,34])
```

### Line Plot

```python
plt.title('Line plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')

plt.plot(x,y, marker ='o', color='blue', mfc='r', ls='--' , ms = 7);

plt.grid()

# marker => < , > , ^ , O , * , H
```

```python
## ls -> line style  , ms -> marker size , mfc -> marker font color
```

```python
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.plot(x,z , marker ='*' , color='red', mfc='b' , ls='--' , ms=10);

plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.plot(x,z , marker ='*' , color='red', mfc='b' , ls='--' , ms=10);
plt.plot(x,y , marker ='^' , color='green', mfc='y' , ls='--' , ms=10);

plt.grid()
```

### Subplot

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.title('Line plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')

plt.plot(x,y, marker ='o', color='blue', mfc='r', ls='--' , ms = 7);

plt.grid()


plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.plot(x,z , marker ='*' , color='red', mfc='b' , ls='--' , ms=10);

plt.grid()
```

### Bar Graph

```python
## To check frequency we use Bar Graph
```

```python
students = [56,65,46,51,61]
courses = ['Java','Python','R','Cotlin','Dart']
```

```python
plt.title('Bar Graph')
plt.xlabel('Courses')
plt.ylabel('Students')

plt.bar(courses,students , color='black', edgecolor='blue', width=0.5, linewidth = 2 , alpha=0.7);

#alpha -> opacity (faded the dark color into light )
```

### Scatter plot

```python
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.scatter(x,y, marker='*', color='r');
plt.grid()
```

### Pie Chart

```python
# pie chart -> gives us the pecentage -> also shows the contribution
```

```python
plt.title('Pie Chart')

plt.pie(students, labels=courses ,
        autopct='%.2f%%', shadow=True ,
        radius=0.8, labeldistance=1.2 ,
        colors=['yellow','skyblue','red','orange','pink'] ,
        startangle=90 , explode = [0,0,0,0.2,0]);
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)

plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,z , marker ='*' , color='red', mfc='b' , ls='--' , ms=10);
plt.plot(x,y , marker ='^' , color='green', mfc='y' , ls='--' , ms=10);
plt.grid()


plt.subplot(2,2,2)

plt.title('Bar Graph')
plt.xlabel('Courses')
plt.ylabel('Students')
plt.bar(courses,students , color='black', edgecolor='blue', width=0.5, linewidth = 2 , alpha=0.7);


plt.subplot(2,2,3)

plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.scatter(x,y, marker='*', color='r');
plt.grid()


plt.subplot(2,2,4)

plt.title('Pie Chart')
plt.pie(students, labels=courses ,
        autopct='%.2f%%', shadow=True ,
        radius=0.8, labeldistance=1.2 ,
        colors=['yellow','skyblue','red','orange','pink'] ,
        startangle=90 , explode = [0,0,0,0.2,0]);


plt.savefig('My Plot')
```

### Box Plot

```python
marks = [67,63,70,71,72,56,51,21,95]
```

```python
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks,75)

q1 = np.percentile(marks,25)
```

```python
q3,q1
```

```python
iqr = q3 - q1
iqr  # interquartile range
```

```python
# Lower Bound
lb = q1 - 1.5 * iqr

# Upper Bound
ub = q3 + 1.5 * iqr
```

```python
lb , ub
```
